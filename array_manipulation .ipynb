```python
def array_manipulation(divisor,array):
    """
    Function to returns a list that is divisiable by any input of any list
    
    Arguments:
    divisor - An Interger value that is dividing each element
    array - Provide a list of values to check
    """
    b=[]
    for i in array:
        if i%divisor ==0:
            b.append(i)
    print(b)
```

```python
a = [6, 7, 8, 4, 2, 9, 3, 5, 7]
```

```python
array_manipulation(3,a)
```
